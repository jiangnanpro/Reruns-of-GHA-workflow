```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels_0 = [
    'Test Flakiness',
    'Branch Discrepancy',
    'External Dependency Issue',
    'Workflow Dependency Issue',
    'Dynamic External Input',
    'Mishandling Edge Cases',
    'GitHub API Rate Limitation',
    'Don’t know/No logs available'
]

labels_1 = [
    'Test Flakiness',
    'External Service Unusable',
    'Authentication Issue',
    'External Dependency Issue',
    'Don’t know/No logs available'
]

sizes_0 = [13, 9, 6, 6, 2, 2, 2, 10]  # Number of occurrences

sizes_1 = [20, 4, 3, 1, 22]  # Number of occurrences

# Define explode to separate the slices a bit
explode_0 = [0.02 for i in range(len(labels_0))]
explode_1 = [0.02 for i in range(len(labels_1))]

# Colors
colors_0 = [
    (231/255, 98/255, 84/255),
    (239/255, 138/255, 71/255),
    (247/255, 170/255, 88/255),
    #(255/255, 208/255, 111/255),
    (255/255, 230/255, 183/255),
    (170/255, 220/255, 224/255),
    (114/255, 188/255, 213/255),
    #(82/255, 149/255, 173/255),
    (55/255, 103/255, 149/255),
    (30/255, 70/255, 110/255)
]

colors_1 = [
    (231/255, 98/255, 84/255),
    (239/255, 138/255, 71/255),
    #(247/255, 170/255, 88/255),
    #(255/255, 208/255, 111/255),
    (255/255, 230/255, 183/255),
    #(170/255, 220/255, 224/255),
    (114/255, 188/255, 213/255),
    #(82/255, 149/255, 173/255),
    (55/255, 103/255, 149/255),
    (30/255, 70/255, 110/255)
]
```

```python
fig, ax = plt.subplots(1,1, figsize=(18, 6))

wedges_0, texts, autotexts_0 = ax.pie(sizes_0, explode=explode_0, labels=None, autopct='%1.1f%%', shadow=True, startangle=90, colors=colors_0)


# Function to adjust text position
def adjust_text_position(texts, offset=0.2):
    for text in texts:
        x, y = text.get_position()
        angle = text.get_rotation()
        # Move the text a bit away from the center by adjusting its x and y coordinates
        text.set_position((x + offset * x, y + offset * y))
        
adjust_text_position(autotexts_0)

ax.legend(wedges_0, labels_0, fontsize=13, loc="center left", bbox_to_anchor=(0.95, -0.25, 0.5, 1))

#ax.set_title('Causes of flakiness in workflow reruns', fontsize = 14, pad=0, loc='center')

plt.setp(autotexts_0, size=9, weight="bold", color="black")

plt.savefig('../plots/piechart_flakiness_workflow.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
